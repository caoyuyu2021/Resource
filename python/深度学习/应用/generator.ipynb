```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("D:\\Jupyter notebook\\data\\python\\iris_dataset\\irisInputs.csv",header=None)
label = pd.read_csv("D:\\Jupyter notebook\\data\\python\\iris_dataset\\irisTargets.csv",header=None)
data = data.values
label = label.values
```

```python
def generator(data, label, batch_size, min_index=None, max_index=None, shuffle=False, drop_remainder=False):
    #为下界设置初始值
    if min_index is None:
        min_index = 0
        
    #为上界设置初始值
    if max_index is None:
        max_index = len(data)
        
    i = min_index
    
    while 1:
        #如果不丢弃数据进行分批
        if drop_remainder:
            #如果需要打乱数据
            if shuffle:
                rows = np.random.randint(i, max_index, size = batch_size)
            else:
                #如果溢界，则重头开始取数
                if i + batch_size <= max_index:
                    rows = np.arange(i, min(i+batch_size, max_index))
                    i += len(rows)
                else:
                    rows = np.arange(i, min(i+batch_size, max_index))
                    i = min_index
                
            samples = np.zeros((len(rows), data.shape[-1]))
            targets = np.zeros((len(rows), ))
            for j, row in enumerate(rows):
                samples[j] = data[row]
                targets[j] = label[row]
            yield samples, targets
            
        #如果丢弃数据进行分批
        else:
            #如果需要打乱顺序
            if shuffle:
                rows = np.random.randint(i, max_index, size = batch_size)
            else:
                #如果溢界，则重头开始取数
                if i + batch_size > max_index:
                    i = min_index
                rows = np.arange(i, min(i+batch_size, max_index))
                i += len(rows)

            samples = np.zeros((len(rows), data.shape[-1]))
            targets = np.zeros((len(rows), ))
            for j, row in enumerate(rows):
                samples[j] = data[row]
                targets[j] = label[row]
            yield samples, targets
```

```python
for (d,l) in generator(data, label, 40):
    print(l)
```
