# 算法原理

多元回归分析预测法，是指通过对两个或两个以上的自变量与一个因变量的相关分析，建立预测模型进行预测的方法。 当自变量与因变量之间存在线性关系时，称为多元线性回归分析。

# 数据准备

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

特征 说明

CRIM 城镇人均犯罪率

ZN 大于25,000平方英尺的地块划分为住宅用地的比例

INDUS 每个城镇非零售业务的比例

CHAS 查尔斯河虚拟变量（如果 = 1则为河; =0则不为河）

NOX 一氧化氮浓度（每千万）

RM 每间住宅的平均房间数

AGE 自住房屋是在1940年之前建造的比例

DIS 到加州五个就业中心的加权距离

RAD 对径向高速公路的可达性指数

TAX 每10,000美元的全价物业税

PTRATIO 城镇的学生与教师比例

B 1000（Bk-0.63）^ 2其中Bk是城镇的黑人的比例

LSTAT 低社会阶层人口比例％

MEDV 以1000美元为单位的自住房屋的中位数价格

```python
#读取数据
data = pd.read_csv("../data/boston_housing_data.csv")
print(data.head())
```

```python
#准备数据
data = data.dropna()
y = data['MEDV']
x = data.drop(['MEDV'],axis=1).astype('float64')

# 训练集、测试集划分
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state= 42)
```

# 模型训练

```python
# 回归模型、参数
model = LinearRegression()

# 训练模型
model.fit(x_train, y_train)

#打印 coef
print('Regression coefficients:', model.coef_)
```

# 模型保存

```python
# 法一
import joblib

# 保存模型
joblib.dump(model, '../outputs/best_models/lr.pkl')

# 加载模型
model = joblib.load('../outputs/best_models/lr.pkl')
```

```python
# 法二
import pickle

with open('../outputs/best_models/lr.pkl', 'wb') as f:
    pickle.dump(model, f)

#读取Model
with open('../outputs/best_models/lr.pkl', 'rb') as f:
    model = pickle.load(f)
```

# 模型预测

```python
prediction = model.predict(x_test)
```

```python
# 计算R2，均方差
r2 = r2_score(y_test, prediction)
mse = np.sqrt(mean_squared_error(y_test, prediction))
```

```python
print("R2为：", r2)
print("MSE为：", mse)
```
