# 算法原理

# 数据准备

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier # 随机森林
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
```

```python
#读取数据
data = pd.read_csv("../data/iris.csv")
print(data.head())
```

```python
#准备数据
data = data.dropna()
y = data['Species'].apply(lambda x: 1 if x == 'setosa' else 2 if x == 'versicolor' else 3)
x = data.drop(['Species'],axis=1).astype('float64')

# 训练集、测试集划分
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state= 42)
```

# 模型训练

```python
# 分类模型、参数
model = RandomForestClassifier(random_state = 0)
param_grid = {'max_depth': [5, 10, 15], 'n_estimators': [10, 50, 100]}

# GridSearchCV优化参数、训练模型
gsearch = GridSearchCV(model, param_grid)
model = gsearch.fit(x_train, y_train)

# 打印最优结果
print('RandomForest params:', model.best_estimator_)
```

```python
forest = RandomForestClassifier(max_depth=5, n_estimators=50, random_state=0)
forest.fit(x_train, y_train)
forest.feature_importances_ # 特征重要性
```

```python
can = {n:v for n,v in zip(data.columns,forest.feature_importances_)}
plt.barh(list(can.keys()),list(can.values()),align = 'center')
plt.xlabel("Feature Importances")
plt.ylabel("Feature")
```

# 模型保存

```python
# 法一
import joblib

# 保存模型
joblib.dump(model, '../test/best_models/rf.pkl')

# 加载模型
model = joblib.load('../test/best_models/rf.pkl')
```

```python
# 法二
import pickle

with open('../test/best_models/rf.pkl', 'wb') as f:
    pickle.dump(model, f)
 
#读取Model
with open('../test/best_models/rf.pkl', 'rb') as f:
    model = pickle.load(f)
```

# 模型预测

```python
prediction = model.predict(x_test)
```

```python
# 计算准确率
acc = accuracy_score(y_test, prediction)
print("acc为：", acc)
```

```python
print(classification_report(y_test, prediction))
```
