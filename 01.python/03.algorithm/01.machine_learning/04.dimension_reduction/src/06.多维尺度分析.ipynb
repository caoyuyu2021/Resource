# 算法原理

Multidimensional scaling，简称MDS，中文翻译成多维尺度分析。其原理是利用成对样本间的相似性，去构建合适的低维空间，使得样本在此空间的距离和在高维空间中的样本间的相似性尽可能的保持一致。我们可以用这种方式来可视化数据分布。

# 数据准备

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import mglearn
from sklearn.manifold import MDS
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import roc_auc_score
import warnings
warnings.filterwarnings("ignore")
```

```python
#读取数据
data = pd.read_csv("../data/breast_cancer.csv")
print(data.head())
```

```python
#准备数据
data = data.dropna()
y = data['label']
x = data.drop(['label'],axis=1).astype('float64')

# 训练集、测试集划分
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state= 42)
```

# 模型训练

```python
model = MDS(n_components=10)
model.fit(x_train)
model.embedding_.shape # 流形嵌入
```

# 模型保存

```python
# 法一
import joblib

# 保存模型
joblib.dump(model, '../outputs/best_models/mds.pkl')

# 加载模型
model = joblib.load('../outputs/best_models/mds.pkl')
```

```python
# 法二
import pickle

with open('../outputs/best_models/mds.pkl', 'wb') as f:
    pickle.dump(model, f)

#读取Model
with open('../outputs/best_models/mds.pkl', 'rb') as f:
    model = pickle.load(f)
```

# 模型预测

```python
# 转换数据
x_new = model.fit_transform(x_train)
```

```python
x_new.shape
```

# 模型应用

```python
# 先对数据降维，再进行分类
mds = MDS(n_components=10)
mds.fit(x_train)
x_new = mds.fit_transform(x_train)

# 分类模型
knn = KNeighborsClassifier()
param_grid = {'n_neighbors': [2, 3, 4]}

#GridSearchCV优化参数、训练模型
gsearch = GridSearchCV(knn, param_grid)
knn = gsearch.fit(x_new, y_train)

#打印最优结果
print('KNN params:', knn.best_estimator_)
```

```python
x_new = mds.fit_transform(x_test)
prediction = knn.predict(x_new)
```

```python
# 计算准确率
acc = accuracy_score(y_test, prediction)
print("acc为：", acc)
```

```python
print(classification_report(y_test, prediction))
```
