```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans#k均值
X,y = make_blobs(random_state=1)
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
print(X)
print(kmeans.predict(X))
```

```python
kmeans.transform(X)
```

```python
kmeans.cluster_centers_
```

```python
mglearn.plots.plot_agglomerative_algorithm()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=3)#凝聚聚类
ass = agg.fit_predict(X)
mglearn.discrete_scatter(X[:,0],X[:,1],ass)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN                  #DBSCAN
X,y = make_moons(n_samples=200,noise=0.05,random_state=0)
scaler = StandardScaler()
scaler.fit(X)
xs = scaler.transform(X)
dbscan = DBSCAN()
clusters = dbscan.fit_predict(xs)
plt.scatter(xs[:,0],xs[:,1],c=clusters,cmap=mglearn.cm2,s=60)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
