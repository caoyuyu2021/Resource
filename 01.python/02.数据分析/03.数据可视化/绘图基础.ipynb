## 用figure()函数画两张图

如果想画两张图则必须使用plt.figure()两次，可以指定num，也可以不指定num，调用两次它会默认加1。

```python
x=[1,2,3]
y1=[1,2,4]
y2=[1,4,8]
 
plt.figure()
plt.plot(x,y1,color = "red",label = "red")
plt.legend()  #无此语句会不显示右下角label
 
plt.figure()
plt.plot(x,y2,color = "green",label = "green")
plt.legend()  #无此语句会不显示右上角label
plt.show()
```
